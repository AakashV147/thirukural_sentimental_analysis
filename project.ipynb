```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/aakashvenkatraman/Documents/thirukural_project/thirukkural_with_sentiments.csv")

if 'Translation' not in df.columns:
    raise KeyError("Column 'Translation' not found. Check CSV structure.")

df = df.dropna(subset=['Translation'])

sns.countplot(data=df, x='sentiment', palette="coolwarm")
plt.title("Sentiment Distribution of Thirukkural")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()

df['text_length'] = df['Translation'].astype(str).apply(len)

plt.figure(figsize=(10, 5))
sns.histplot(df['text_length'], bins=30, kde=True, color='purple')
plt.title("Distribution of Translation Lengths")
plt.xlabel("Number of Characters")
plt.ylabel("Frequency")
plt.show()

```
